## Аналитическая часть

### Интегральная сумма:

$$\prec f=3^{t-1},\quad t=x+1,\quad 0\le t\le 2$$

$$\sigma_{\tau}(f,\xi) = \sum_{k=1}^{n}3^{\frac{2k}{n}-1}\cdot \frac{2}{n}=\frac{2}{3n}\sum_{k=1}^{n}3^{\frac{2k}{n}}=\frac{2}{3n}\cdot\frac{3^{\frac{2}{n}}(3^{2}-1)}{3^{\frac{2}{n}}-1}$$

### Нахождение предела интегральной суммы:

$$\int_{-1}^{1}3^{x}dx=\frac{1}{3}\int_{0}^{2}3^{t}dt=\lim_{n \to +\infty} \frac{2}{3n}\cdot\frac{3^{\frac{2}{n}}(3^{2}-1)}{3^{\frac{2}{n}}-1} = \frac{8}{3} \cdot\lim_{n \to +\infty} \frac{2}{n}\cdot\frac{3^{\frac{2}{n}}}{\frac{2}{n}\ln3} =\frac{8}{3} \cdot\lim_{n \to +\infty}\frac{3^{\frac{2}{n}}}{\ln3} = \frac{8}{3\ln3}$$

### Значение интеграла по формуле Ньютона-Лейбница:

$$\int_{-1}^{1}3^{x}dx = {F(x)}\bigg|_{-1}^1 = F(1) - F(-1) = {\frac{3^{x}}{\ln3}}\bigg|_{-1}^1 = \frac{3}{\ln3} - \frac{1}{3\ln3} = \frac{8}{3\ln3}$$

$$\frac{8}{3\ln3} = \frac{8}{3\ln3}$$

### Доказательство существования интеграла

Для доказательства интегрируемости функции воспользуемся критерием Римана: $$\forall \varepsilon > 0 \ \exists  \tau:  \ S_\tau(f) - s_\tau(f) < \varepsilon,$$ $$$$ где $$\ S_\tau(f)=\sum_{k=1}^{n}M_{k}\cdot \Delta x_{k}, \quad M_i = \sup\limits_{x \in \Delta_k} f(x), \quad k \in \{1, 2, ..., n\},$$ $$
s_{\tau}(f) = \sum\limits_{i = 1}^nm_i \Delta x_i, \quad m_i = \inf\limits_{x \in \Delta_i} f(x), \quad i \in \{1, 2, ..., n\},$$

Тогда $$\ S_\tau-s_{\tau}=\sum_{k=1}^{n}(M_{k}-m_{k})\cdot\Delta x_{k}⊜,$$ где из монотонности функции $3^x$: $$M_{k} = f(x_{k})$$ $$m_{k}=f(x_{k-1})$$ для  $\Delta _{k} = [x_{k-1}, x_{k}]$

$$⊜ \sum_{k=1}^{n}(f(x_{k})-f(x_{k-1}))\cdot\Delta x_{k}=\sum_{k=1}^{n}\Delta f_{k}\cdot\Delta x_{k}\lt \lambda\sum_{k=1}^{n}\Delta f_{k},$$ $$$$ где $\lambda(\tau)=\max\limits_{k \in \{1, 2, \ \ldots, n \}} \Delta x_k = \frac{\varepsilon}{f(b)-f(a)}, \quad \sum_{k=1}^{n}\Delta f_{k}=f(b)-f(a)$

Получаем $$\lambda\sum_{k=1}^{n}\Delta f_{k}=\lambda(f(b)-f(a))<\varepsilon.$$ Значит данная функция интегрируема

## Программа на Python

```python
from matplotlib import pyplot as plt
import numpy as np
```

### Код основной программы

```python
def integral_sum(func, n, equip, a, b): # func - функция, equip - оснащение [a,b] - данный промежуток, n - число точек разбиения
    dx = (b - a) / n # длина отрезков
    if equip == 'l': # левое
        total = sum([f(a + (k * dx)) for k in range(0, n)])
    elif equip == 'r': # правое
        total = sum([f(a + (k * dx)) for k in range(1, n + 1)])
    else: # среднее
        total = sum([f(a + ((2 * k - 1) / 2 * dx)) for k in range(1, n + 1)])
    return dx * total


def f(x): # данная функция
    return 3 ** x

a,b = [-1, 1] # промежуток
n = 10 # количество точек разбиения
equip = 'l' # разбиение

print('Ответ:', integral_sum(f, n, equip, a, b))
```

## Погрешности

### Вывод формул

Рассмотрим отрезок $[x_{i}; x_{i+1}]$. По формуле Тейлора с остатком в форме Лагранжа: $$f(x)=f(x_{i})+f'(x_{i})(x-x_{i})+\frac{1}{2}f''(\xi_i)(x-x_i)^2, \quad где \space \xi_i\in [x_i;x_{i+1}]$$ Проинтегрируем обе части: $$\int_{x_i}^{x_{i+1}}f(x)dx=\int_{x_i}^{x_{i+1}}f(x_i)dx + \int_{x_i}^{x_{i+1}}f'(x_i)(x-x_i)dx+\int_{x_i}^{x_{i+1}}\frac{1}{2}f''(\xi_i)(x-x_i)^2dx$$ $$$$ $$\int_{x_i}^{x_{i+1}}f(x)dx=f(x_i)\cdot\Delta x_i+f'(x_i)\cdot\frac{\Delta x_i^2}{2}+\frac{1}{6}f''(\xi_i)\Delta x_i^3$$

Тогда для отрезка [-1;1]:
$$\int_{-1}^{x_{1}}3^xdx=\sum_{i=0}^{n-1}f(x_i)\cdot\Delta x_i+f'(x_i)\cdot\frac{\Delta x_i^2}{2}+\frac{1}{6}f''(\xi_i)\Delta x_i^3,\quad \Delta x_i=\frac{b-a}{n}$$
